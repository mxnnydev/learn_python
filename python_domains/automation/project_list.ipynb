## 🗓️ 6-Week Python Automation Mastery Roadmap

**Assumes: Python Fundamentals ✅ | Goal: Build real tools, fast**

---

### 🧱 **Week 1: File System Automation**

#### 🛠️ Projects:

1. **Bulk File Renamer**
2. **Desktop Cleaner Bot**

#### 🔑 Skills:

* `os`, `shutil`, `pathlib`
* File paths, loops, string formatting

#### 🧠 Milestone:

* You can manipulate folders & files programmatically like a true hacker.

---

### ⏰ **Week 2: Time-Based & Scheduled Tasks**

#### 🛠️ Projects:

3. **Daily Reminder Bot**
4. **Time Tracker Tool**

#### 🔑 Skills:

* `datetime`, `time`, `schedule`
* Loops with delays, writing logs

#### 💡 Bonus:

* Automate a motivational popup or quote every morning.

---

### 📤 **Week 3: Email Automation & Notifications**

#### 🛠️ Projects:

5. **Auto Email Sender**
6. **Email + Quote of the Day Bot**

#### 🔑 Skills:

* `smtplib`, `email.message`, env vars for passwords
* Basic `csv`/`txt` integration

#### 🧠 Milestone:

* You can send templated emails on a schedule from your own Python script.

---

### 📈 **Week 4: Data Cleanup & Reporting**

#### 🛠️ Projects:

7. **CSV Data Cleaner**
8. **Excel Automation Tool**

#### 🔑 Skills:

* `csv`, `pandas`, `openpyxl`
* Read, filter, clean, export data

#### 💡 Bonus:

* Auto-clean data for lead lists or contact exports

---

### 📑 **Week 5: PDF & Text Processing**

#### 🛠️ Projects:

9. **PDF Invoice Extractor**
10. **Regex Email/Phone Extractor**

#### 🔑 Skills:

* `PyPDF2`, `pdfplumber`, `re`
* Parse, match, extract from documents

#### 🧠 Milestone:

* You can scan & extract key info from messy PDFs and text files.

---

### 🌐 **Week 6: Web Automation & Scraping**

#### 🛠️ Projects:

11. **News Headline Scraper**
12. **Price Tracker Tool**

#### 🔑 Skills:

* `requests`, `BeautifulSoup`, saving to `CSV`
* Headers, selectors, user-agent basics

#### 💡 Bonus:

* Auto-send price drops to email

---

## ✅ Graduation Challenge (Optional Final Week)

**⚡ Build your own automation stack** that:

* Scrapes → Cleans → Sends Email
  *E.g. Get top AI news headlines daily → save to file → auto-email to yourself*

---

## 🧠 System Notes:

* 📂 Store each project in its own folder (`mkdir Week_1_BulkRenamer`)
* 📝 Keep a `README.md` inside each with what it does and what you learned
* 🔁 Every 2 weeks, **refactor** 1–2 projects with better code quality or new libraries
